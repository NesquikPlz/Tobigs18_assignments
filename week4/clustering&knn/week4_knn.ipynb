###데이터 불러오기

```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/dataset/BlackFriday.csv")
```

```python
data
```

범주형 데이터, 이상치 및 결측치 처리

```python
for i in data.columns:
    print(f"{i} :{len(data[str(i)].unique())} {data[str(i)].unique()}")

#해당 결과를 보면 범주형을 int형으로 바꿔줘야 할 column이 Age와 Stay_In_Current_City_Years 인 것을 알 수 있다
#또한 Gender와 City_Category에는 one-hot encoding을 적용시켜주겠다
```

```python
data['Age'] = data['Age'].replace({'0-17' : 9, '18-25' : 22, '26-35' : 31, '36-45' : 41, '46-50' : 48, '51-55' : 53, '55+' : 60 }) # 나이범위를 평균값(int)으로 바꿔준다
```

```python
data['Stay_In_Current_City_Years'] = data['Stay_In_Current_City_Years'].replace({'4+':5})
data = data.astype({'Stay_In_Current_City_Years':'int'})
```

```python
#원핫인코딩
pd.get_dummies(data = data, columns = ['Gender', 'City_Category'], prefix = ['Gender', 'City_Category'])
```

```python
#결측치 확인
data.isnull().sum()
```

```python
#Product_Category_2와 Product_Category_3의 결측치가 매우 많으므로 두 칼럼을 삭제해주겠다. 또한 purchase와  Product_Category_1이 결측치인 경우는 해당 row만 삭제해주겠다

data = data.drop(columns = ['User_ID', 'Product_ID', 'Gender', 'City_Category', 'Product_Category_2', 'Product_Category_3']) #불필요한 칼럼 함께 삭제
data = data.dropna()
data
```

```python
data.info()
```

###데이터 분리 및 스케일링

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1],data['Purchase'], test_size = 0.2, random_state = 41)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

###knn 학습

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()
knr.fit(X_train_scaled, y_train)
```

```python
print(knr.score(X_test_scaled, y_test))
#hyperparameter tuning 없이 학습했을 때 정확도가 0.55 가량인 것을 확인할 수 있다
```

###Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV

leaf_size = list(range(20,50))
n_neighbors = list(range(1,15))
weights = ['uniform', 'distance']
p=[1,2]

#dictionary type으로 변환해준다
hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p, weights = weights)

knr2 = KNeighborsRegressor()

clf = GridSearchCV(knr2, hyperparameters, cv=5)

best_model = clf.fit(X_train_scaled,y_train)

print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
```

```python
knr = KNeighborsRegressor(leaf_size=29, n_neighbors=14, p = 1, weights='distance')
best_model = knr.fit(X_train_scaled,y_train)
```

```python
print(knr.score(X_test_scaled, y_test))
#조금이지만 hyperparameter tuning을 통해 accuracy가 올라간 것을 확인할 수 있다.
```
